```python
from dotenv import load_dotenv

load_dotenv() # Load the API keys from .env file
from quantrl_lab.data.loaders.alpaca_loader import AlpacaDataLoader
```

```python
# Initialize the Interface to Alpaca Data Loader
alpaca_client = AlpacaDataLoader()
```

```python
# Fetch historical data for a specific stock
symbol = "AAPL"  # Example: Apple Inc.
start_date = "2025-01-01"
end_date = "2025-06-01"
historical_data = alpaca_client.get_historical_ohlcv_data(symbol, start_date, end_date)
# Print the fetched historical data
historical_data.head()
```

```python
# Fetch news data for a specific stock
news_data = alpaca_client.get_news_data(symbol, start="2025-06-01", end="2025-06-02", limit=50)
# Print the fetched news data
news_data.head()
```

```python
# Fetch latest quotes for a specific stock
latest_quotes = alpaca_client.get_latest_quote(symbol)
# Print the latest quotes
print(latest_quotes)
```

```python
# Fetch latest trades for a specific stock
latest_trades = alpaca_client.get_latest_trade(symbol)
# Print the latest quotes
print(latest_trades)
```
